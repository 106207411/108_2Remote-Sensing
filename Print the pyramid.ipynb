```python
"""n is an integer and 0<n<16 """

n=int(input('Please enter a number of layers:'))
data=[2**i for i in range(n)]
space='{:>4}'.format(' ')
for r in range(1,n+1):
    print(space*(n-r),end='')
    for i in data[:r]:
        print(f'{i:>4}',end='')
    for i in data[:r-1][::-1]:
        print(f'{i:>4}',end='')
    print()
```
