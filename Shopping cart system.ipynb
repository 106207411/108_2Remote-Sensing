```python
product = {"1":"banana", "2":"apple", "3":"orange", "4":"pineapple", "5":"tangerine"}
cart = {}
print("shopping list: 1.banana, 2.apple, 3.orange, 4.pineapple, 5.tangerine")
while True:
    step = input("Please enter the instruction:\n1.add items 2.remove items q.quit b.Foot the bill")
    if step == "1":
        add = input("Please enter the id of added item:")
        if add in product:
            cart[add] = product[add]
            print("Added successfully！Now you have", cart)
    if step == "2":
        remove = input("Please enter the id of removed item:")
        if remove in product:
            del cart[remove]
            print("Removed successfully！Now you have", cart)
    elif step == "q":
        break
    elif step == "b":
        print("Purchasing succesfully!", cart)
        break
```
