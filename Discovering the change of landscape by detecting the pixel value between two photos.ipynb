```python
from PIL import Image, ImageDraw

def compare_pix(im0,im1,i,j):
    pix_im0 = im0.getpixel((i,j)) # return the pixel value of (R, G, B)
    pix_im1 = im1.getpixel((i,j))
    global dif_r, dif_g, dif_b, dif_avg
    dif_r = pix_im0[0] - pix_im1[0]
    dif_g = pix_im0[1] - pix_im1[1]
    dif_b = pix_im0[2] - pix_im1[2]
    dif_avg = abs(round((dif_r + dif_g + dif_b) / 3))
    dif_r = abs(dif_r)
    dif_g = abs(dif_g)
    dif_b = abs(dif_b)

    if dif_r < threshold_r :
        dif_r = 0
    elif dif_g < threshold_g :
        dif_g = 0
    elif dif_b < threshold_b :
        dif_b = 0
    elif dif_avg < threshold_avg :
        dif_avg = 0

def main():
    before_img = input("The old picture:")
    After_img = input("The new picture:")    
    im0 = Image.open(before_img)
    im1 = Image.open(After_img)
    width = im0.size[0]
    height = im0.size[1]
    im_r = Image.new('RGB',(width, height))
    im_g = Image.new('RGB',(width, height))
    im_b = Image.new('RGB',(width, height))
    im_rgb = Image.new('RGB',(width, height))
    im_avg = Image.new('RGB',(width, height))    
    draw_r = ImageDraw.Draw(im_r)
    draw_g = ImageDraw.Draw(im_g)
    draw_b = ImageDraw.Draw(im_b)
    draw_rgb = ImageDraw.Draw(im_rgb)
    draw_avg = ImageDraw.Draw(im_avg)
    
    global threshold_r, threshold_g, threshold_b, threshold_avg
    threshold_r = int(input("threshold_r:")) # set up an arbitrary value for threshold
    threshold_g = int(input("threshold_g:"))
    threshold_b = int(input("threshold_b:"))
    threshold_avg = int(input("threshold_greyscale:"))
    
    for i in range(width):
        for j in range(height):
            compare_pix(im0,im1,i,j)
            draw_r.point((i,j), fill = (dif_r, 0 ,0))
            draw_g.point((i,j), fill = (0, dif_g ,0))
            draw_b.point((i,j), fill = (0, 0, dif_b))
            draw_rgb.point((i,j), fill = (dif_r, dif_g, dif_b))
            draw_avg.point((i,j), fill = (dif_avg, dif_avg, dif_avg)) # greyscale image
            
    im_r.save("compare_r.jpg")
    im_g.save("compare_g.jpg")
    im_b.save("compare_b.jpg")
    im_rgb.save("compare_rgb.jpg")
    im_avg.save("compare_avg.jpg")
    
if __name__ == '__main__':
    main()
```
